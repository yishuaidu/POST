```python
df = pd.read_csv('alldata.csv') 
df = df[df.banner_pos <= 6 ]
df = df[df.site_id <= 4734 ]
df = df[df.app_id <= 8551 ]
df = df[df.device_model <= 8247 ]
print(max(df.banner_pos.unique()))
print(max(df.site_id.unique()))
print(max(df.app_id.unique()))
print(max(df.device_model.unique()))
len(df)
```

```python
df0 = df[df.click == 0]
df0 = df0.reset_index()
df1 = df[df.click == 1]
df1 = df1.reset_index()
```

```python
import random


t1 = np.random.choice(28000, replace=False, size=28000)
test1 = df0.iloc[t1]
train1 = df0.iloc[list(set(df0.index.tolist()) - set(t1))]

t2 = np.random.choice(28000, replace=False, size=28000)
test2 = df1.iloc[t2]
train2 = df1.iloc[list(set(df1.index.tolist()) - set(t2))]

tr = train1.append(train2)
tr = tr.reset_index()
test = test1.append(test2)
test = test.reset_index()
```
